```python
from transformers import pipeline
nlp = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')
samples = [
    'I love this workshop!',
    'This is a challenging problem but exciting to solve.'
]
print(nlp(samples))
```
